```python
import os
import pandas as pd
from openpyxl import load_workbook

folder_path = r'S:\BigPrecip\Precip\U_Mad\ALL TX PRECIP\TX56.xlsx'
combined_df = pd.DataFrame(columns=['STATION_ID', 'NAME', 'DATE', 'TIME', 'LAT', 'LON', 'ELEV', 'PRECIP'])  # Initialize an empty DataFrame with the desired columns

if os.path.isfile(folder_path) and folder_path.endswith('.xlsx'):
    try:
        workbook = load_workbook(folder_path, read_only=True, data_only=True)
        sheets = workbook.sheetnames

        for sheet_name in sheets:
            sheet = workbook[sheet_name]
            data = sheet.values
            columns = next(data)
            df = pd.DataFrame(data, columns=columns)
            
            df['DATETIME_GMT'] = df['DATETIME_GMT'].astype(str)  # Convert to string
            df[['DATE', 'TIME']] = df['DATETIME_GMT'].str.split(' ', expand=True)
            df.drop('DATETIME_GMT', axis=1, inplace=True)
            df.columns = ['STATION_ID', 'PRECIP', 'DATE', 'TIME']

            site_info = pd.read_excel(r'S:\BigPrecip\Precip\U_Mad\all_texas_sites.xlsx')
            df_merged = pd.merge(df, site_info[['SiteID', 'LatitudeDD', 'LongitudeDD', 'Elevation (feet)', 'SiteName']], left_on='STATION_ID', right_on='SiteID', how='left')
            
            df_reordered = df_merged[['STATION_ID', 'SiteName', 'DATE', 'TIME', 'LatitudeDD', 'LongitudeDD', 'Elevation (feet)', 'PRECIP']]
            df_reordered.columns = ['STATION_ID', 'NAME', 'DATE', 'TIME', 'LAT', 'LON', 'ELEV', 'PRECIP']
            
            combined_df = pd.concat([combined_df, df_reordered], ignore_index=True)
            
    except Exception as e:
        print(f"Error reading file '{folder_path}': {str(e)}")

    combined_df['DATE'] = pd.to_datetime(combined_df['DATE'])  # Convert date column to datetime format
    combined_df['DATE'] = combined_df['DATE'].dt.strftime('%Y-%m-%d')  # Format the date column as "YYYY-MM-DD"
    
    filename = os.path.basename(folder_path)
    filename = os.path.splitext(filename)[0]  # Remove the file extension from the filename
    new_filepath = os.path.join(r'C:\Pretty Precip Data\U_Mad _data _done', filename + '.csv')  # Set .csv extension explicitly
    combined_df.to_csv(new_filepath, encoding='utf-8', index=False)
        
    print(f"Processed file: {filename}")

```
